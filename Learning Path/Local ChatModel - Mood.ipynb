```python
# LLM from Ollama
local_model = "mistral"
llm = ChatOllama(model=local_model)
```

```python
QUERY_PROMPT = PromptTemplate(
    input_variables=["mood"],
    template="""What activity should I do if I'm feeling {mood}?""",)
```

```python
prompt = ChatPromptTemplate.from_template(QUERY_PROMPT.template)
```

```python
chain = (
    prompt
    | llm
    | StrOutputParser()
)
```

```python
chain.invoke(input(""))
```
